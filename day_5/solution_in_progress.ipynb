### Parse Moves

```python
# Parse the move lines
def parse_move(move_str: str) -> List[int]:
    """Parses a single move string into a list of three integers.
    
    move_str is a single move line, e.g. 'move x from y to z'.
    Function returns move_ints, a list of 3 integers.
    """

    # Change every other word to an int; save to a list.  
    move_ints = [int(item) for item
                   in move_str.split(' ')[1::2]]  
    
    return move_ints
```
