```python
move_codes = {
    'A': 'rock',
    'B': 'paper',
    'C': 'scissors',
    'X': 'rock',
    'Y': 'paper',
    'Z': 'scissors',
    'rock': 1,
    'paper': 2,
    'scissors': 3
}

loss = 0
tie = 3
win = 6
```

```python
input_file = 'input.txt'

with open(input_file, 'r') as f:
    rounds = f.readlines()

for round in rounds:
    moves = round.split()
    opponent_move = move_codes[moves[0]]
    my_move = move_codes[moves[1]]

    round_score = 0
    if my_move == opponent_move:
        round_score += tie
    elif opponent_move == 'rock':
        if my_move == 'paper':
            round_score += win
        elif my_move == 'scissors':
            round_score += loss
    elif opponent_move == 'paper':
        if my_move == 'scissors':
            round_score += win
        elif my_move == 'rock':
            round_score += loss
    else:
        if my_move == 'rock':
            round_score += win
        elif my_move == 'paper':
            round_score += loss

    round_score += move_codes[my_move]
     
    print(round_score)
```
